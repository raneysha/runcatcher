```python
import psycopg
from deepface import DeepFace
import os
from tqdm import tqdm
import pandas as pd
import json
```

```python
URL_DATABASE = 'postgresql://postgres:pass@localhost:5432/deepface'
```

```python
con = psycopg.connect(URL_DATABASE)
```

```python
instances = []
for dirpath, dirnames, filenames in os.walk("uploads/"):
  for filename in tqdm(filenames):
    if filename.endswith(".jpg") or filename.endswith(".png"):
      instances.append(f"{dirpath}/{filename}")
```

```python
instances
```

```python
try:
  for img_path in tqdm(instances):
    DeepFace.register(
      img = img_path,
      l2_normalize = True,
      connection = con
    )
except Exception as e:
  print(f"Error processing {img_path}: {e}")
```

```python
dfs = DeepFace.search(
  img = "2.png",
  l2_normalize = True,
  connection = con
)
```

```python
result_json = json.dumps(dfs[0].to_dict(orient='records'), indent=2)
print(result_json)
```
